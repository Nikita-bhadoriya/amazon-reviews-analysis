```python
plt.figure(figsize=(6,4))
sns.countplot(x='reviews.rating',data=data1)
plt.title("Rating Distribution")
plt.show()
```

```python
data1['Year']=data1['reviews.date'].dt.year
data1['Year'].value_counts().sort_index().plot(kind='bar',figsize=(8,4),title='Reviews per year')
```

```python
#  Top 10 products by number of reviews
data1['name'].value_counts().head(10)
```

```python
# Average rating per product
avg_rating=data1.groupby('name')['reviews.rating'].mean().sort_values(ascending=False)
avg_rating.head(10)
```

```python
# Most helpful reviews
data1.sort_values('reviews.numHelpful',ascending=False)[['name','reviews.text','reviews.numHelpful']].head(10)
```

```python
# Review Length
data1['review_length']=data1['reviews.text'].apply(lambda x:len(str(x).split()))
```

```python
data1[['reviews.text', 'review_length']].head()
```

```python
def categorize_rating(r):
    if r>= 4:
        return 'Good'
    elif r==3:
        return 'Neutral'
    else:
        return 'Bad'
data1['rating_category']=data1['reviews.rating'].apply(categorize_rating)
```

```python
data1[['reviews.rating', 'rating_category']].head(10)
```

```python
data1.groupby('Year')['reviews.rating'].mean().plot(
    kind='line', title='Average Rating Over Time', figsize=(8,4))
```
